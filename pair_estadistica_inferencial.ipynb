2- Ejercicios intervalo de confianza. Al igual que en el caso del ejercicio de la clase invertida, piensa un motivo por el que sería interesante calcular estos intervalos de confianza para la extracción de insights.

```python
df = pd.read_csv("files/world_data_full_apply_nulos.csv", index_col = 0 ) 
df.head()
```

- Intervalo de Confianza para la Esperanza de Vida   
    - Contexto: Trabajas en una organización de salud global y deseas estimar el intervalo de confianza al 95% para la esperanza de vida promedio de cada continente.   
    - Nota: Cuidado porque puede que os salgan algunos nulos, genera un código para que no salgan nulos en los intervalos de confianza.



```python
for continent, data in df.groupby('continent')['life_expectancy']:
    media_esp = data.mean()
    error_esp = stats.sem(data, nan_policy= "omit")
    nivel_confianza_esp = 0.95
    libertad_esp = len(data) - 1
    valor_critico_esp = stats.t.ppf((1 + nivel_confianza_esp)/2, df=libertad_esp)

    int_inferior = media_esp - valor_critico_esp * error_esp
    int_superior = media_esp + valor_critico_esp * error_esp

    print(f"Intervalo de Confianza para la Esperanza de Vida de {continent.upper()}")
    print(f"Media Muestral: {np.round(media_esp, 2)}")
    print(f"Error Estándar: {np.round(error_esp, 2)}")
    print(f"Nivel de Confianza: {nivel_confianza_esp}")
    print(f"Valor Crítico: {np.round(valor_critico_esp, 2)}")
    print(f"Intervalo de Confianza: ({np.round(int_inferior, 2)}, {np.round(int_superior, 2)})")
    print('---')
```

- Intervalo de Confianza para el PIB   
    - Contexto: Eres una economista y quieres calcular el intervalo de confianza al 90% para el Producto Interno Bruto (PIB) de los continentes.

```python
for continent, data in df.groupby('continent')['gdp']:
    media_gdp = data.mean()
    error_gdp = stats.sem(data, nan_policy= "omit")
    nivel_confianza_gdp = 0.90
    libertad_gdp = len(data) - 1
    valor_critico_gdp = stats.t.ppf((1 + nivel_confianza_gdp)/2, df=libertad_gdp)

    int_inferior = media_gdp - valor_critico_gdp * error_gdp
    int_superior = media_gdp + valor_critico_gdp * error_gdp

    print(f"Intervalo de Confianza para el PIB de {continent.upper()}")
    print(f"Media Muestral: {np.round(media_gdp, 2)}")
    print(f"Error Estándar: {np.round(error_gdp, 2)}")
    print(f"Nivel de Confianza: {nivel_confianza_gdp}")
    print(f"Valor Crítico: {np.round(valor_critico_gdp, 2)}")
    print(f"Intervalo de Confianza: ({np.round(int_inferior, 2)}, {np.round(int_superior, 2)})")
    print('---')
```

- Intervalo de Confianza para la Tasa de Natalidad   
    - Contexto: Trabajas en una agencia gubernamental que se ocupa de cuestiones demográficas y deseas determinar el intervalo de confianza al 99% para la tasa de natalidad promedio en Asia.

```python
for continent, data in df.groupby('continent')['birth_rate']:
    if continent == 'Asia':
        media_birth = data.mean()
        error_birth = stats.sem(data, nan_policy= "omit")
        nivel_confianza_birth = 0.99
        libertad_birth = len(data) - 1
        valor_critico_birth = stats.t.ppf((1 + nivel_confianza_birth)/2, df=libertad_birth)

        int_inferior = media_birth - valor_critico_birth * error_birth
        int_superior = media_birth + valor_critico_birth * error_birth

        print(f"Intervalo de Confianza para el PIB de {continent.upper()}")
        print(f"Media Muestral: {np.round(media_birth, 2)}")
        print(f"Error Estándar: {np.round(error_birth, 2)}")
        print(f"Nivel de Confianza: {nivel_confianza_birth}")
        print(f"Valor Crítico: {np.round(valor_critico_birth, 2)}")
        print(f"Intervalo de Confianza: ({np.round(int_inferior, 2)}, {np.round(int_superior, 2)})")
        print('---')
```

- Intervalo de Confianza para la Tasa de Desempleo   
    - Contexto: Eres un analista de políticas públicas y deseas calcular el intervalo de confianza al 95% para la tasa de desempleo promedio en los países de Oceanía.


```python
for continent, data in df.groupby('continent')['unemployment_rate']:
    if continent == 'Oceania':
        media = data.mean()
        error = stats.sem(data, nan_policy= "omit")
        nivel_confianza = 0.95
        libertad = len(data) - 1
        valor_critico = stats.t.ppf((1 + nivel_confianza)/2, df=libertad)

        int_inferior = media - valor_critico * error
        int_superior = media + valor_critico * error

        print(f"Intervalo de Confianza para el PIB de {continent.upper()}")
        print(f"Media Muestral: {np.round(media, 2)}")
        print(f"Error Estándar: {np.round(error, 2)}")
        print(f"Nivel de Confianza: {nivel_confianza}")
        print(f"Valor Crítico: {np.round(valor_critico, 2)}")
        print(f"Intervalo de Confianza: ({np.round(int_inferior, 2)}, {np.round(int_superior, 2)})")
        print('---')
```
